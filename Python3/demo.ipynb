# Bar Plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
c=['r',"g","y"]
plt.xlabel("language" ,fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("wscubetech",fontsize=15)
plt.bar(x,y,color=c)
plt.show()
```
